```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuration of the plots we will create

# read in the data

df = pd.read_csv('/Users/patrycjapidek/Downloads/movies.csv')
#print("the data set has {} rows and {} columns".format(df.shape(0),df.shape(1)))
```

```python
# let's look at the data
df.head(10)
```

```python
# sorting by the gross

df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# drop any duplicates
df['name'].drop_duplicates().head()
```

```python
# what things are most correlated to the gross value(testing correlations)
# budget high correlation
# company will have high correlation 
```

```python
# scatterplot with budget vs gross

plt.scatter(x= df['budget'], y= df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
# plot the budget vs gross using seaborn 

sns.regplot(x = 'budget', y = 'gross', data= df, scatter_kws = {"color":"red"}, line_kws = {"color":"green"})
```

```python
# let's start looking at the correlation #pearson is used by default

df.corr() # only works on numerical fields
# We can see that there is high correlation between budget and gross
```

```python
df.corr(method = "kendall")
```

```python

df.corr(method = "spearman")
```

```python
correlation_matrix = df.corr(method = "pearson") # the lighter/brighter colour - more correlation
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# looking at company ( not numeric), but we can create representation ( unique identifier)

df_numerized = df
for col_name in df_numerized:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes # random number representation
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = "pearson") # the lighter/brighter colour - more correlation
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
#using unstacking to quickly locate correlation

correlation_mat = df_numerized.corr()
corr_pairs =correlation_mat.unstack()
corr_pairs.head()
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) >0.5]
high_corr
```
